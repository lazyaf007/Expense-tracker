```python
import pandas as pd

# Load data
df = pd.read_csv("expenses_project.csv")

# Convert date
df['Date'] = pd.to_datetime(df['Date'])

# Create Month column
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.strftime('%B')

# Total spending
total_spend = df['Amount'].sum()
print("Total Spending:", total_spend)

# Category-wise spending
category_spend = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
print(category_spend)

# Monthly spending
monthly_spend = df.groupby('Month_Name')['Amount'].sum()
print(monthly_spend)

# Payment method analysis
payment_spend = df.groupby('Payment_Mode')['Amount'].sum()
print(payment_spend)

# Highest expense day
max_expense = df.loc[df['Amount'].idxmax()]
print(max_expense)

# Export cleaned data
df.to_csv("cleaned_expenses.csv", index=False)
```
